```python
from matplotlib import pyplot as plt
import numpy as np

from deeptime.clustering import BoxDiscretization
from deeptime.markov import TransitionCountEstimator
from deeptime.markov.msm import MaximumLikelihoodMSM
from deeptime.data import prinz_potential
```

```python
# Get synthetic 2D data, This will be replaced with tICA result
prinz_system = prinz_potential()

# Visualize input data
xs = np.linspace(-1., 1., 500)
ys = prinz_system.potential(xs)
plt.plot(xs, ys)
plt.xlabel('x')
plt.ylabel('V(x)');
```

```python
# Simulate trajectories for 50 particles.
data = prinz_system.trajectory(np.zeros((50, 1)), 20000)
print(f"Output shape {data.shape} for {data.shape[0]} particles, {data.shape[1]} frames "
      f"in {data.shape[2]} dimension.")
```

```python
# See where particles stayed. If E well is deep, more particles will stay longer, causing higher y value.
# Total number of data points: 50 particles * 20K frames = 1M. (If you integrate(Sum, because it's discrete) this graph, it's 1M)
plt.hist(np.concatenate(data), bins=150);
```
